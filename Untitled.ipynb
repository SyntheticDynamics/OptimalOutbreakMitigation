The purpose of this notebook is to provide some code that will result in some information regarding a particular system with particular stuff. First of all, we will need to import some modules and set some variables.

```python
import sir
import matplotlib.pyplot as plt
```

```python
### Set your parameters. The class always takes imax and umax, but can handle either beta/gamma or r0. Set the ones you
### wish to use, and leave the rest as None
imax = 0.1
umax = 0.5
gamma = 0.2
beta = 0.5
r0 = None

if r0 is None:
    A = sir.SIR()
    A.set_params([imax, umax, gamma, beta], flag = "bg")
elif beta is None and gamma is None:
    A = sir.SIR()
    A.set_params([imax, umax, r0], flag = "r")
```

```python
### Set an initial point for your system, get important features, and find the trajectory of least time to the safe zone
s0, i0 = [0.8, 0.06]
P = A.add_point(s0, i0)
A.find_curves()
A.find_regions()
A.get_shortest()
T = P.least_time
```

```python
fig, ax = plt.subplots()
ax.set_xlim(A.sbar, 1)
ax.set_xlabel(r"$S$")
ax.set_ylim(0, A.imax*1.1)
ax.set_ylabel(r"$I$")
ax.plot(A.tau.s, A.tau.i, color = "xkcd:blue", linestyle = "-")
ax.plot(A.phi.s, A.phi.i, color = "xkcd:red", linestyle = "-")
ax.plot([A.sbar, 1], [A.imax, A.imax], color = "xkcd:black", linestyle = ":")
ax.plot(T.s, T.i, color = "xkcd:vomit green", linestyle = "-")
```
